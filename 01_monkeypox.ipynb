### 1. What is Monkeypox (Mpox) ?
- Mpox is a viral infectious disease caused by the monkeypox virus, a member of the Orthopoxvirus genus in the Poxviridae familiy-the same family as the virus that causes smallpox.
The disease was first identified in 1958 in monkeys used for research, which led to its original name. The first human case was recorded in 1970 in the Democratic Republic of the Congo. 
- [Symptom] Mpox typically causes a combination of flu-like symptoms and a distinctive rash. Common symptoms include:
    - fever
    - headache
    - muscle aches
    - back pain
    - low energy
    - swollen lymph nodes
    - rash or mucosal lesions that last 2-4 weeks
- [Transmission] Mpox is a zoonotic disease, meaning it can spread from animals to humans and between humans.
- Human-to-human transmission occurs mainly through close contact, including:   
    - skin-to-skin contact(sexual contact)
    - mouth-to-mouth (kissing)
    - contact with contaminated objects like clothing or bedding.
- [Treatment and prevention] There is no specific treatment for mpox; care focuses on relieving symptoms and preventing complications. Antiviral medications like tecovirimat may be used in severe cases, but their effectiveness is still being studied. Vaccines developed for smallpox have been found to be effective agains mpox, and vaccination is recommended for people at high risk or those recently exposed.

```python
def unique_data(data) : 
    for col in data.columns :
        print(col, len(data[col].unique()), data[col].dtype, data[col].unique())
```

- 위 함수의 목적은 데이터프레임 내 각 변수의 특성(고유값 분포와 데이터 유형)을 빠르게 탐색하기 위한 EDA 도구임.

- `len(data[col].unique)` : 고유값의 개수
- `data[col].dtype` : 데이터 타입
